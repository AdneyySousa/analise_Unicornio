```python
#chamando data
base['Date Joined']
```

```python
#convertendo a coluna de datas de objeto para formato de datas
base['Date Joined'] = pd.to_datetime(base['Date Joined'])
base['Date Joined'].head()
```

```python
#extraindo ano e mes da coluna de datas
#criando coluna na base chamada mes
base['Mes'] = pd.DatetimeIndex(base['Date Joined']).month

#criando coluna e passando ano
base['Ano'] = pd.DatetimeIndex(base['Date Joined']).year
base.head()
```

```python

#criando uma tabela analitica
analise_agrupada = base.groupby(by=['Country','Ano','Mes','Company','Valuation ($B)']).count()['Unnamed: 0'].reset_index()
```

```python
#localizando o mes e ano, onde o Brasil se tornou unicornio dentro da tabela
analise_agrupada.loc[
    analise_agrupada['Country'] == 'Brazil'
]
```

```python
#formatando uma coluna dentro da tabela
pd.to_numeric(base['Valuation ($B)'].apply(lambda Linha : Linha.replace('$','')))
```

```python
#inserindo dentro da coluna valuation, a formatação dos valores
base['Valuation ($B)'] = pd.to_numeric(base['Valuation ($B)'].apply(lambda Linha : Linha.replace('$','')))
```

```python
base.head()
```

```python
#fazendo uma analise encima do valor
analise_pais = base.groupby(by=['Country']).sum()['Valuation ($B)'].reset_index()
#mostrando tabela na tela
analise_pais.head()
```

```python
#ordenando valores da tabela
analise_valor = analise_pais.sort_values('Valuation ($B)',ascending= False)

analise_valor.head()
```

```python
#passando os valores para dentro de um grafico
plt.figure(figsize=(15,9))
plt.title('analise do valor por País')
plt.plot(analise_valor['Country'],analise_valor['Valuation ($B)'])
plt.xticks(rotation=45,ha='right');
```
